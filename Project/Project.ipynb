```python
import numpy as np
import matplotlib.pyplot as plt
from reservoirpy.nodes import Reservoir, Ridge

# Generate sine wave data
T = 1000
t = np.linspace(0, 10*np.pi, T)
data = np.sin(t)

# Prepare training and testing datasets
inputData = data[:-1].reshape(-1, 1)
target = data[1:].reshape(-1, 1)

trainLen = 800
xTrain, yTrain = inputData[:trainLen], target[:trainLen]
xTest, yTest = inputData[trainLen:], target[trainLen:]

# Define the Echo State Network
reservoir = Reservoir(units=200, lr=0.3, sr=0.9, input_scaling=0.5)
readout = Ridge(ridge=1e-6)

esn = reservoir >> readout

# Train the network
esn = esn.fit(xTrain, yTrain)

# Predict
yPred = esn.run(xTest)

# Plot results
fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(yTest, label="True Signal")
ax.plot(yPred, label="Predicted Signal", linestyle="--")
ax.legend()
ax.set_title("Sine Wave Prediction with ReservoirPy")
```

```python
from reservoirpy.datasets import lorenz
from reservoirpy.nodes import Reservoir, Ridge
import numpy as np
import matplotlib.pyplot as plt

# Load Lorenz data
timeseries = lorenz(10000) # from example
data = timeseries[:, 0].reshape(-1, 1)

trainLen = 8000
xTrain = data[:trainLen]
yTrain = data[1:trainLen + 1]
xTest = data[trainLen:-1]
yTest = data[trainLen + 1:]

# Define ESN NO noise
reservoir = Reservoir(units=300, sr=1.25, lr=0.3, input_scaling=0.5)
readout = Ridge(ridge=1e-6)

esn = reservoir >> readout

# Train and predict
esn = esn.fit(xTrain, yTrain)
yPred = esn.run(xTest)

# Plot
fig, ax = plt.subplots(figsize =(12,4))
ax.plot(yTest[:1000], label = "True")
ax.plot(yPred[:1000], label="Predicted", linestyle="--")
ax.set_title("Lorenz Attractor Prediction (x-axis)")
ax.legend()
```
